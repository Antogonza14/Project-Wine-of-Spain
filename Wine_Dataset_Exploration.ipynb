```python
# Aqui selecciono los vinos de mi tierra que reflejan la diversidad y la riqueza de la Región de Murcia 
# Conocida por producir vinos con gran carácter y calidad 
# Filtro los datos para obtener solo los vinos de la región de "Murcia Jumilla"

jumilla_wines = data[data['region'] == 'Jumilla']
jumilla_wines_list = jumilla_wines[['winery', 'wine', 'year']].reset_index(drop=True)
jumilla_wines_list
```

```python
import pandas as pd
# Filtrar los vinos de las regiones de Yecla de la variedad de uva Monastrell
vinos_murcia = data[data['region'].isin(['Yecla'])]

print(vinos_murcia)
```

```python
# Los 10 vinos más caros (datos usados para despues hacer los graficos).
most_expensive_wines = data.nlargest(10, 'price')[['winery', 'wine', 'year', 'price']].reset_index(drop=True)

# Los 10 vinos con mejores calificaciones (datos usados para despues hacer los graficos).
top_rated_wines = data.nlargest(10, 'rating')[['winery', 'wine', 'year', 'rating']].reset_index(drop=True)

most_expensive_wines, top_rated_wines
```

```python
# Es un gráfico que muestra las relaciones entre pares de variables en un conjunto de datos 
# Colorea los puntos según las calificaciones (rating). 
# En la diagonal, muestra la distribución de cada variable. 
# Fuera de la diagonal, muestra cómo cada par de variables se relaciona entre sí. 
# El color indica cómo estas relaciones varían con diferentes calificaciones. 
# Ayudando a identificar patrones y correlaciones entre variables y calificaciones de manera visual e intuitiva.
# Muy interesante como punto de partida.

sns.pairplot(data,hue='rating')
```

```python
#Gráfico muestra relación entre precio y calificación de vinos de Jumilla, incluyendo tendencia aparentemente al alza.

plt.figure(figsize=(12, 8))
sns.scatterplot(x='price', y='rating', data=jumilla_wines, color="purple", alpha=0.7, s=100)
sns.regplot(x='price', y='rating', data=jumilla_wines, scatter=False, color="gold")
plt.title('Calificación vs. Precio de los vinos de Jumilla', fontsize=20)
plt.xlabel('Precio', fontsize=16)
plt.ylabel('Calificación', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.grid(True, which='major', linestyle='--', linewidth='0.5', color='grey')
plt.tight_layout()

plt.show()
```

```python
# Selecciono datos de una bodega concreta con varios años disponibles
# Vega Sicilia es una buena candidata en mi opinion por su representación en los datos
vega_sicilia_prices = data[data['winery'] == 'Vega Sicilia'].groupby('year')['price'].mean().reset_index()

plt.figure(figsize=(10, 2))
plt.plot(vega_sicilia_prices['year'], vega_sicilia_prices['price'], marker='', linestyle='-', color='red')
plt.title('Tendencia de Precios de Vega Sicilia (Media Anual)')
plt.xticks([])  
plt.yticks([])
plt.box(False) 

plt.show()
```
